##Correlation matrix

```python
import seaborn as sn
sns.set(rc={'figure.figsize':(11.7,8.27)})
sns.set_style('darkgrid')

corrMatrix = df.corr()
sn.heatmap(corrMatrix, annot=True)
plt.show()
```

##**Regplot**

```python
fig, ax = plt.subplots(nrows=2, ncols= 3,figsize= (20,15))
sns.regplot(ax=ax[0,0], x ='trip_distance', y='total_amount', data =df_copy,color= 'lightseagreen',marker='o')
sns.regplot(ax=ax[0,1], x ='fare_amount', y='total_amount', data =df_copy,color= 'lightseagreen',marker='o')
sns.regplot(ax=ax[0,2], x ='tip_amount', y='total_amount', data =df_copy,color= 'lightseagreen',marker='o')
sns.regplot(ax=ax[1,0], x='tolls_amount',y='total_amount',data=df_copy,color= 'lightseagreen',marker='o')
sns.regplot(ax=ax[1,1], x ='duration', y='total_amount', data =df_copy,color= 'lightseagreen',marker='o')
sns.regplot(ax=ax[1,2], x ='total_amount', y='total_amount', data =df_copy,color= 'lightseagreen',marker='o')
```
